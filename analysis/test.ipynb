```python
from figure1 import extract_slha
```

```python
import pandas as pd
df = pd.DataFrame(extract_slha("M2from0to1200M15000mu1000and5000"))
```

```python
df.head()
```

```python
df_1000 = df[df['mu'] == 1000].sort_values(by='M_2')
df_5000 = df[df['mu'] == 5000].sort_values(by='M_2')


df_1000 = df_1000[(df_1000['M_2'] % 5 == 0) & (df_1000['M_2'] >100)]
df_5000 = df_5000[(df_5000['M_2'] % 5 == 0) & (df_5000['M_2'] >100)]
```

```python
df_5000.head()
```

```python
import matplotlib.pyplot as plt
from figure1 import plot2

plot2([df_5000['BR(N2->N1+Z)'], df_5000['BR(N2->N1+ff)'], df_5000['BR(N2->N1+h)']], df_5000['M_2'], 100, 5000, 10, ['BR(N2->N1+Z)', 'BR(N2->N1+ff)', 'BR(N2->N1+h)'], 'Branching ratio in the neutralino 2 decay')
```

```python
plot2([df_5000['BR(C1->N1+W)'], df_5000['BR(C1->N1+ff)']], df_5000['M_2'], 100, 5000, 10, ['BR(C1->N1+W)', 'BR(C1->N1+ff)'], 'Branching ratio in the Chargino 1 decay')
```

```python
from figure1 import plot2_ATLAS_style
from figure1 import plot2_ATLAS_style_double
plot2_ATLAS_style([df_5000['BR(C1->N1+W)'], df_5000['BR(C1->N1+ff)']], df_5000['M_2'], 100, 5000, 10, [r'$\tilde{\chi}_1^\pm \, \to\, W+\tilde{\chi}_1^0$', r'$\tilde{\chi}_1^0\, \to\,  W+\tilde{\chi}_1^0$'], r'wino-bino scenario', labelpos='right', yaxis=r"$\tilde{\chi}_1^\pm$", textposy=0.6, save='C1')
```

```python
plot2_ATLAS_style([df_5000['BR(N2->N1+h)'], df_5000['BR(N2->N1+Z)'], df_5000['BR(N2->N1+ff)']], df_5000['M_2'], 100, 5000, 10, [r'$\tilde{\chi}_2^0 \, \to\, h+\tilde{\chi}_1^0$',r'$\tilde{\chi}_2^0 \, \to\, Z+\tilde{\chi}_1^0$',  r'$\tilde{\chi}_2^0\, \to\, Z^*+ \tilde{\chi}_1^0$'], r'wino-bino scenario', labelpos='right', yaxis=r"$\tilde{\chi}_2^0$", textposy=0.6, save='N2')
```

```python
liste_1000 = [df_1000['BR(N2->N1+h)'], df_1000['BR(N2->N1+Z)']]
plot2_ATLAS_style_double([df_5000['BR(N2->N1+h)'], df_5000['BR(N2->N1+Z)']],liste_1000, df_5000['M_2'], 100, 5000, 1000, 10, [r'$\tilde{\chi}_2^0 \, \to\, h+\tilde{\chi}_1^0$',r'$\tilde{\chi}_2^0 \, \to\, Z+\tilde{\chi}_1^0$'], r'wino-bino scenario', labelpos='right', yaxis=r"$\tilde{\chi}_2^0$", textposy=0.58, save='N2_both')
```
